# Visualize loss

A notebook for loss visualization

```python
def parse_training_output_file_and_return_epochs_and_losses_for_each_epoch(path_to_training_output_file):
    # I assume a very strict output format, so I'm going to rely heavily on "where things are supposed to be"
    # and use indicies
    training_output_file = open(path_to_training_output_file, "r")
    training_output_file_all_lines = training_output_file.readlines()
    
    epochs = []
    losses_for_each_epoch = []
    
    EPOCH_NUMBER_LINE_INDEX_BEGIN = 14
    LOSS_NUMBER_LINE_INDEX_BEGIN = 6
    LOSS_NUMBER_LINE_INDEX_END = 14
    
    for line in training_output_file_all_lines:
        if "Epoch number" in line:
            epoch_number = int(line[EPOCH_NUMBER_LINE_INDEX_BEGIN:])
            epochs.append(epoch_number)
        if "[    6/    7]" in line:
            loss = float(line[LOSS_NUMBER_LINE_INDEX_BEGIN:LOSS_NUMBER_LINE_INDEX_END])
            losses_for_each_epoch.append(loss)
            
    return epochs, losses_for_each_epoch
```

```python
PATH_TO_TRAINING_OUTPUT_LOG = "../mini_train_training_output_log-old.txt"
```

```python
epochs, losses = parse_training_output_file_and_return_epochs_and_losses_for_each_epoch(PATH_TO_TRAINING_OUTPUT_LOG)
```

Let's visualize the training loss:

```python
import matplotlib.pyplot

matplotlib.pyplot.plot(epochs, losses)
```
